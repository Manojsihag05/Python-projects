```python
df['Year']=df['Date'].dt.year
df_2022 = df[df['Year']==2022]
monthly_sales_profit_2022 = df_2022.groupby(['Year',df_2022['Date'].dt.month])[['Sales','Profit']].sum()
print(monthly_sales_profit_2022)
```

```python
#Which are the Top 20 drugs by sales of 2022
drug_sales_2022 = df[df['Year'] == 2022].groupby('Drug Name')['Sales'].sum()
top_20_drugs_2022 = drug_sales_2022.sort_values(ascending=False).head(20)
print(top_20_drugs_2022)
```
